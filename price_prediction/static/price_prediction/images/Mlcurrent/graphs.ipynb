```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("makegraph.csv", low_memory=False)
df.head()
```

```python
tomato_data = df[df['Commodity'] == 'Potato White']

# Group data by season and calculate the average price
seasonal_avg_prices = tomato_data.groupby('Season')['Average'].mean()

# Plotting
plt.figure(figsize=(8, 6))
seasonal_avg_prices.plot(kind='bar', color='skyblue')
plt.title('Average Price Fluctuation of Potato White by Season')
plt.xlabel('Season')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.savefig('bar2.png')
plt.show()
```

```python
import plotly.express as px

# Assuming your DataFrame is named df and contains necessary columns

# Filter data for Tomato Big(Nepali)
tomato_data = df[df['Commodity'] == 'Tomato Big(Nepali)']

# Group data by date and calculate the average price
date_avg_prices = tomato_data.groupby('Date')['Average'].mean().reset_index()

# Create an interactive line graph using Plotly
fig = px.line(date_avg_prices, x='Date', y='Average', title='Price Fluctuation of Tomato Big(Nepali)',
              labels={'Date': 'Date', 'Average': 'Average Price'})

fig.update_traces(mode='lines+markers', hovertemplate='Date: %{x}<br>Price: %{y:.2f}')

fig.update_layout(hovermode='x unified', xaxis=dict(showspikes=True, spikedash='dot', spikethickness=1))

# Save the figure as an image file (PNG format)
fig.write_image('tomato_price_fluctuation.png')
```
