```python
import pandas as pd
from datetime import datetime, timedelta
from ipywidgets import interact
from IPython.display import display


file_path = 'preprocessed_data.csv'  # Adjust the file path as needed
selected_data = pd.read_csv(file_path)

# Assuming 'selected_commodities' is a list of commodities you want to analyze
selected_commodities = ['Onion Dry (Indian)', 'Tomato Big(Nepali)', 'Potato White', 'Carrot(Local)', 'Apple(Jholey)', 'Spinach', 'Banana', 'Cucumber(Local)', 'Garlic Dry Nepali', 'Lettuce']
days_to_look_back = 7

def plot_live_table(selected_commodities):
    # Filter data for the selected commodities
    selected_data_filtered = selected_data[selected_data['Commodity'].isin(selected_commodities)]

    # Calculate the start and end date for the last week
    end_date = datetime.today().strftime('%Y-%m-%d')
    start_date = (datetime.today() - timedelta(days=days_to_look_back)).strftime('%Y-%m-%d')

    # Filter data for the last week
    last_week_data = selected_data_filtered[(selected_data_filtered['Date'] >= start_date) & (selected_data_filtered['Date'] <= end_date)]

    # Calculate the percentage change in the 'Average' column for the last week
    last_week_data['Percentage Change'] = last_week_data.groupby('Commodity')['Average'].pct_change() * 100

    # Drop rows with NaN values in 'Percentage Change'
    last_week_data.dropna(subset=['Percentage Change'], inplace=True)

    # Pivot the DataFrame for better display
    table_data = last_week_data.pivot(index='Commodity', columns='Date', values='Percentage Change')

    # Display the table
    display(table_data)

# Execute the function with initial commodities
plot_live_table(selected_commodities)
```

```python
import pandas as pd
from datetime import datetime, timedelta
from ipywidgets import interact, SelectMultiple, Button
from IPython.display import display

selected_commodities = ['Onion Dry (Indian)', 'Tomato Big(Nepali)', 'Potato White', 'Carrot(Local)', 'Apple(Jholey)', 'Spinach', 'Banana', 'Cucumber(Local)', 'Garlic Dry Nepali', 'Lettuce']
days_to_look_back = 7

select_commodities = SelectMultiple(options=selected_commodities, value=[selected_commodities[0]], description='Commodities')

# Create a button to trigger the update
button = Button(description="Update Table")

# Display the widgets
display(select_commodities, button)


def plot_live_table(selected_commodities):
    # Convert the variable to a list
    selected_commodities = list(selected_commodities)
    
    # Create an empty DataFrame to store the results
    result_table = pd.DataFrame()

    for selected_commodity in selected_commodities:
        # Filter data for the selected commodity
        selected_data_filtered = selected_data[selected_data['Commodity'] == selected_commodity].copy()

        # Calculate the percentage change in the 'Average' column
        selected_data_filtered['Percentage Change'] = selected_data_filtered['Average'].pct_change(fill_method=None) * 100

        # Drop rows with NaN values in 'Percentage Change'
        selected_data_filtered.dropna(subset=['Percentage Change'], inplace=True)

        # Display the resulting DataFrame for the current commodity
        print(f"\nResults for {selected_commodity}:\n")
        display(selected_data_filtered[['Commodity', 'Percentage Change']])

        # Append the results to the overall result table
        result_table = pd.concat([result_table, selected_data_filtered[['Commodity', 'Percentage Change']]])

    # Pivot the overall result table for better display
    table_data = result_table.pivot_table(index='Commodity', values='Percentage Change', aggfunc='mean')

    # Display the overall table
    print("\nOverall Result Table:\n")
    display(table_data)

# Callback function to handle button click
def on_button_click(button):
    selected_commodities = select_commodities.value
    plot_live_table(selected_commodities)

# Attach the callback function to the button
button.on_click(on_button_click)
```
